```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
df = pd.read_csv("C:/Users/thota/OneDrive/ドキュメント/house-price-prediction/data/processed/cleaned_house_data.csv")
```

### price distribution

```python
plt.figure()
sns.histplot(df["SalePrice"], kde = True)
plt.title("SalePrice Distribution")
plt.show()
```

```python
num_df = df.select_dtypes(exclude=["object"])
corr = num_df.corr(numeric_only=True)["SalePrice"].sort_values(ascending=False)
```

```python
plt.figure(figsize=(8, 6))
corr.head(15).plot(kind="bar")
plt.title("Top 15 Numeric Correlations with SalePrice")
plt.ylabel("Correlation")
plt.show()
```

### 3) Boxplot of a strong categorical feature (Neighborhood is common)

```python
if "Neighborhood" in df.columns:
    plt.figure(figsize=(12, 16))
    top_n = df["Neighborhood"].value_counts().head(15).index
    sns.boxplot(data=df[df["Neighborhood"].isin(top_n)], x="Neighborhood", y="SalePrice")
    plt.xticks(rotation=45, ha="right")
    plt.title("Salesprice by Neighborhood (Top 15 by count)")
    plt.show()
```
